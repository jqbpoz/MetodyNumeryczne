```julia
using Plots
```

```julia
N=1000
h=0.01
y= ones(N);
h_power2 = h*h
y[1] = 1
y[2] = -4/(h_power2-6)
y[3] = 16/(h_power2-6)+3
for i in 3:N-1 
    y[i+1] = (2-h_power2)*y[i] - y[i-1]
end
x_values = 1:N;
y_values = [y[n] for n in x_values]
plot(x_values*h, y_values,xlabel="nh", ylabel="y_n", legend=:bottomright)


```
